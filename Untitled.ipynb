```python
#read in CSV files
percap2018="percapincomemetro2018.csv"
percap2021="percapincome2021bls.csv"
priceparities="priceparities2018.csv"
geofips="geocodes.csv"
```

```python
#make it into dataframes
pc2018=pd.read_csv(percap2018, skiprows=4,thousands=',')
pc2018.columns
```

```python
#income 2018 as table, clarify column names, drop unnecessary columns
pc2018=pc2018.rename(columns={"Unnamed: 0":"Area","2017": "2017 PerCapIncome","2018":"2018 PerCapIncome"})
pc2018.drop(pc2018.columns[[3,4,5,6]],axis=1,inplace=True)
pc2018.head()
```

```python
#income 2021 as table
pc2021=pd.read_csv(percap2021,skiprows=3,thousands=',')
pc2021.head()
```

```python
#Clarify columns 
pc2021=pc2021.rename(columns={"Unnamed: 0":"Area","2019": "2019 PerCapIncome","2020":"2020 PerCapIncome",
                       "2021":"2021 PerCapIncome"})
pc2021.at[2,"Area"]= 'United States nonmetropolitan portion'
pc2021.drop(pc2021.columns[[4,5,6,7]],axis=1,inplace=True)
pc2021.head()
```

```python
#compared cost of living to national average as table
priceparity=pd.read_csv(priceparities,skiprows=3)
priceparity.head()
```

```python
#Clarify columns
priceparity=priceparity.rename(columns={"Unnamed: 0":"Area","Unnamed: 1": "All Items 2018",
                                        "Unnamed: 2":"Goods", "Rents":"Rents 2018"})
priceparity.drop(priceparity.columns[[2,4]],axis=1,inplace=True)
priceparity.at[1,"Area"]= 'United States'
priceparity.head(15)
```

```python
#Add in list of GeoFips Codes
codes=pd.read_csv(geofips)
codes.head()
```

```python
#Clarify columns
codes=codes.rename(columns={"msa":"GeoFIPs ID","name_msa": "Area"})
codes.drop(codes.columns[[2,3,4]],axis=1,inplace=True)
codes.head()
```

```python
#merge income tables together
income=pd.merge(pc2018,pc2021, how ="left" , on=["Area"])
#include income change - first change datatype to numbers 
#income.apply(pd.to_numeric,errors='ignore')
income['2021 PerCapIncome'] = income['2021 PerCapIncome'].astype(float)
income['2017 PerCapIncome'] =income['2017 PerCapIncome'].astype(float)
#math it
income["5 year $ Increase"]=income.apply(lambda x: x["2021 PerCapIncome"]-x["2017 PerCapIncome"], axis=1)
income["5 year % change"]=income.apply(lambda x: ((x["2021 PerCapIncome"]-x["2017 PerCapIncome"])/x["2017 PerCapIncome"])*100, axis=1)
#pretty up the number format
pd.options.display.float_format='{:,.2f}'.format
income.head(10)
```

```python
#A function for linear regression
def line_regress(x_values, y_values, title,label_location):
    #math it
    (slope, intercept, rvalue, pvalue, stderr)=linregress(x_values,y_values) 
    regress_values = x_values * slope + intercept
    equation="y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    #plot it
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(equation,label_location,fontsize=15,color="red")
    plt.ylabel(title)
    plt.savefig("income.png")
    plt.show()
    print (f"R value is equal to {rvalue**2}")
```

```python
y_values=new_cities["Millennial Population Change as Percentage Growth"]
x_values=new_cities["5 year % change"]
plt.xlabel("Income Growth %")
line_regress(x_values,y_values,"Population Growth and Rise in Income",(18,15))
```

```python
y_values=new_cities["Millennial Population Change as Percentage Growth"]
x_values=new_cities["3 year Change"]
plt.xlabel("Dollar Parity")
line_regress(x_values,y_values,"Population Growth and Cost of Living",(-2,3))
```
